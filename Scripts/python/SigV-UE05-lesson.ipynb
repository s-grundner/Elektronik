```python
import sympy as sy
import scipy.signal as sc
import numpy as np
import matplotlib.pyplot as plt
```

## Aufgabe 1

```python
z, a, b, c = sy.symbols('z a b c')
Xz = z/(z-a) + z/(z-b) + (c+z)/z
Xz = sy.factor(Xz).collect(z)
Xz
```

```python
sy.denom(Xz).expand()
```

## Aufgabe 3

```python
num, den = [1/4, 1/2, 1], [1, 1/2, 1/4]
w, mag, phase = sc.dbode(sc.TransferFunction(num, den, dt=0.0001))
zeros, poles, gain = sc.tf2zpk(num, den)
```

```python
omega = np.linspace(0, 2 * np.pi, 100)
fig, ax = plt.subplots()
ax.scatter(zeros.real, zeros.imag, marker='o')
ax.scatter(poles.real, poles.imag, marker='x')
ax.plot(np.cos(omega), np.sin(omega), linestyle="--", color='black')
ax.grid(True)
ax.axis('equal')
```

```python
fig, axs = plt.subplots(2, 1)
axs[0].semilogx(w, mag)
axs[1].semilogx(w, phase)
```
