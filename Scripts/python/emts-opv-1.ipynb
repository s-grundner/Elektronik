```python
import matplotlib.pyplot as plt
import scipy as sp

def plt_bode(w, mag, phase):
	# subplots
	fig, (ax1, ax2) = plt.subplots(2, 1)
	ax1.semilogx(w, mag)
	ax1.set_title('Bode Plot')
	ax1.set_ylabel('Magnitude (dB)')
	ax1.grid(which='both', axis='both')
	ax2.semilogx(w, phase)
	ax2.set_ylabel('Phase (degrees)')
	ax2.set_xlabel('Frequency (rad/s)')
	ax2.grid(which='both', axis='both')
	plt.show()
```

```python
r1 = 1e3
r2 = 400
c = 300e-9

norm = 1+r1/r2

sys = sp.signal.lti([c, 1/r1 + 1/r2], [c, 1/r1])
w, mag, phase = sys.bode()

sys_norm = sp.signal.lti([c, (1/r1 + 1/r2)], [c*norm, norm/r1])
w_norm, mag_norm, phase_norm = sys_norm.bode()

# plt_bode(w, mag, phase)
plt_bode(w_norm, mag_norm, phase_norm)
```

```python
cx = 150e-9
cf = 450e-9
rf = 1e3
sys = sp.signal.lti([-cx, 0], [cf, 1/rf])
w, mag, phase = sys.bode()

plt_bode(w, mag, phase)
```
